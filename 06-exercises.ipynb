## 05. 下列有關 F-Score 的敘述何者正確？

\begin{equation}
F_{\beta} = (1 + \beta^2) \cdot \frac{precision \cdot recall}{(\beta^2 \cdot precision) + recall}

4. F-Score 在 $\beta = 1$ 的時候稱為 F1-score。
5. F-Score 指標愈高，代表模型的表現愈好。
