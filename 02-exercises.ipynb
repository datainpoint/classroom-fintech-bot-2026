```python
import unittest

def run_suite(test_class):
    suite = unittest.TestLoader().loadTestsFromTestCase(test_class)
    runner = unittest.TextTestRunner(verbosity=2)
    test_results = runner.run(suite)
    number_of_failures = len(test_results.failures)
    number_of_errors = len(test_results.errors)
    number_of_test_runs = test_results.testsRun
    number_of_successes = number_of_test_runs - (number_of_failures + number_of_errors)
    chapter_percentage = number_of_successes * 100 / number_of_test_runs
    print("你的練習題完成率為 ... {:.2f}% ({}/{})".format(chapter_percentage, number_of_successes, number_of_test_runs))

class TestDayTwo(unittest.TestCase):
    def test_01(self):
        self.assertEqual(answer_01, 4)
    def test_02(self):
        self.assertEqual(answer_02, 2)
    def test_03(self):
        self.assertEqual(answer_03, 3)
    def test_04_convert_km_to_mile(self):
        self.assertTrue(convert_km_to_mile(42.195) > 26)
        self.assertTrue(convert_km_to_mile(42.195) < 27)
        self.assertTrue(convert_km_to_mile(21.095) > 13)
        self.assertTrue(convert_km_to_mile(21.095) < 14)
    def test_05_convert_fahrenheit_to_celsius(self):
        self.assertTrue(convert_fahrenheit_to_celsius(212) >= 100.0)
        self.assertTrue(convert_fahrenheit_to_celsius(32) >= 0.0)
    def test_06_show_integer_with_commas_and_digits(self):
        self.assertEqual(show_integer_with_commas_and_digits(1000), "1,000.00")
        self.assertEqual(show_integer_with_commas_and_digits(10000), "10,000.00")
        self.assertEqual(show_integer_with_commas_and_digits(100000), "100,000.00")
        self.assertEqual(show_integer_with_commas_and_digits(1000000), "1,000,000.00")
        self.assertEqual(show_integer_with_commas_and_digits(10000000), "10,000,000.00")
    def test_07_is_a_divisor(self):
        self.assertFalse(is_a_divisor(2, 3))
        self.assertFalse(is_a_divisor(3, 4))
        self.assertTrue(is_a_divisor(1, 3))
        self.assertTrue(is_a_divisor(3, 3))
        self.assertTrue(is_a_divisor(1, 4))
        self.assertTrue(is_a_divisor(2, 4))
        self.assertTrue(is_a_divisor(4, 4))
    def test_08_retrieve_the_first_and_last_element(self):
        the_first_and_last_element = retrieve_the_first_and_last_element([2, 3, 5, 7, 11])
        self.assertEqual(the_first_and_last_element["first"], 2)
        self.assertEqual(the_first_and_last_element["last"], 11)
        the_first_and_last_element = retrieve_the_first_and_last_element([2, 3, 5])
        self.assertEqual(the_first_and_last_element["first"], 2)
        self.assertEqual(the_first_and_last_element["last"], 5)
        the_first_and_last_element = retrieve_the_first_and_last_element(["Python", "Reticulate", "Anaconda"])
        self.assertEqual(the_first_and_last_element["first"], "Python")
        self.assertEqual(the_first_and_last_element["last"], "Anaconda")
        the_first_and_last_element = retrieve_the_first_and_last_element(["Python", "Reticulate", "Anaconda", "Skywalker"])
        self.assertEqual(the_first_and_last_element["first"], "Python")
        self.assertEqual(the_first_and_last_element["last"], "Skywalker")
    def test_09_find_taipei_citys_zip_code(self):
        self.assertEqual(find_taipei_citys_zip_code("中正區"), 100)
        self.assertEqual(find_taipei_citys_zip_code("大同區"), 103)
        self.assertEqual(find_taipei_citys_zip_code("中山區"), 104)
        self.assertEqual(find_taipei_citys_zip_code("松山區"), 105)
        self.assertEqual(find_taipei_citys_zip_code("大安區"), 106)
        self.assertEqual(find_taipei_citys_zip_code("萬華區"), 108)
        self.assertEqual(find_taipei_citys_zip_code("信義區"), 110)
        self.assertEqual(find_taipei_citys_zip_code("內湖區"), 114)
        self.assertEqual(find_taipei_citys_zip_code("南港區"), 115)
        self.assertEqual(find_taipei_citys_zip_code("北投區"), 112)
        self.assertEqual(find_taipei_citys_zip_code("士林區"), 111)
        self.assertEqual(find_taipei_citys_zip_code("文山區"), 116)
    def test_10_check_data_type(self):
        self.assertEqual(check_data_type(0), 'int')
        self.assertEqual(check_data_type(1.0), 'float')
        self.assertEqual(check_data_type(False), 'bool')
        self.assertEqual(check_data_type(True), 'bool')
        self.assertEqual(check_data_type('5566'), 'str')
        self.assertEqual(check_data_type(None), 'NoneType')
    
run_suite(TestDayTwo)
```
